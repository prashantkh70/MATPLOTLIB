## Seaborn - Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic,x="sex")  #countplot - shows each category count using bars  
plt.show()                           #data= "dataframe" , on x-axis ="columns name"  , y-axis="column name"
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic,y="sex")  #countplot - shows each category count using bars  
plt.show()                           #data= "dataframe" , on x-axis ="columns name"  , y-axis="column name"
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")      # seaborn style
sns.countplot(data=titanic,x="sex",hue="pclass")    # hue- splits up data in side the column mention on x or y.
plt.show() 
```

## Categorical plots

### Stripplot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.stripplot(data=titanic,x="sex",y="age", jitter=False,hue=None,dodge=False)  #stripplot-   
plt.show()                 #jitter-True -points and they overlap, so that it is easier to see the distribution  
                        #dodge- When using ``hue`` nesting, setting this to ``True`` will separate the strips 
                                #for different hue levels along the categorical axis. Otherwise, the points for 
                                #each level will be plotted on top of each other.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.stripplot(data=titanic,x="sex",y="age", jitter=True,hue=None,dodge=False)  
plt.show() 
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.stripplot(data=titanic,x="sex",y="age", jitter=True,hue='pclass',dodge=False)  
plt.show()                               # It is hard to differentiate the pclass points, so dodge should be True
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.stripplot(data=titanic,x="sex",y="age", jitter=True,hue='pclass',dodge=True)  
plt.show()                                          # dodge =True seperated the points from pclass and make it clear.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.stripplot(data=titanic,x="sex",y="age", jitter=True,hue='survived',dodge=True)  
plt.show() 
```

### Swarmplot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.swarmplot(data=titanic,x="sex",y="age",hue='pclass',dodge=True)  #swarmplot- locates all dot side by side.   
plt.show()                    # also shows frequency distribution of the groups
```

### violine plot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)        # violineplot- better shows the distribution. The line shown by plot is known as 
sns.violinplot(data=titanic,x="sex",y="age",hue='pclass',dodge=True)    #'kernal desnity estimate'. 
plt.show()      # It is kind of smooth histogram having frequency distribution.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)        # violineplot- better shows the distribution. The line shown by plot is known as 
sns.violinplot(data=titanic,x="sex",y="age",hue='pclass',dodge=True)    #'kernal desnity estimate'. 
sns.swarmplot(data=titanic,x="sex",y="age",hue='pclass',dodge=True,color='black')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.violinplot(data=titanic,x="pclass",y="age",hue='sex',dodge=True,split=False)  #stripplot-   
plt.show() 
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.violinplot(data=titanic,x="pclass",y="age",hue='sex',dodge=True,split=True)  #split= True
plt.show() 
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.barplot(data=titanic,x="pclass",y="age",hue='sex',dodge=True) 
plt.show()  
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)      
sns.pointplot(data=titanic,x="pclass",y="age",hue='sex',dodge=True) #pointplot-straight line shows confideance interval
plt.show()  
```

## JOINTPLOTS/ REGRESSION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
sns.set(font_scale=1.5)                                          # shows histogram for both paramaeters
sns.jointplot(data=titanic,x="age",y="fare",height=8,kind="scatter")
plt.show()
```

```python
sns.set(font_scale=1.5)                                          #reg- shows regression line for both paramaeters
sns.jointplot(data=titanic,x="age",y="fare",height=8,kind="reg")
plt.show()
```

```python
sns.set(font_scale=1.5)                                          # darker the area ,majority of points are there. 
sns.jointplot(data=titanic,x="age",y="fare",height=8,kind="kde")
plt.show()
```

```python
sns.set(font_scale=1.5)                                          # 
sns.jointplot(data=titanic,x="age",y="fare",height=8,kind="hex")
plt.show()
```

### lmplot()

```python
sns.set(font_scale=1.5)                                  #shows confidence interval along with regression line.
sns.lmplot(data=titanic,x="age",y="fare",aspect=1,height=8,hue=None)   # hue- used to split in two columns.
plt.show()
```

```python
sns.set(font_scale=1.5)                                  #shows confidence interval along with regression line.
sns.lmplot(data=titanic,x="age",y="fare",aspect=1,height=8,hue="sex")   # hue- used to split in two columns.
plt.show()
```

```python
sns.set(font_scale=1.5)                                  #shows confidence interval along with regression line.
sns.lmplot(data=titanic,x="age",y="fare",aspect=1,height=8,col="sex")   # col- to get seperate columns.
plt.show()
```

```python
sns.set(font_scale=1.5)                                  #shows confidence interval along with regression line.
sns.lmplot(data=titanic,x="age",y="fare",aspect=1,height=8,row="sex")   # row- to get seperate rows.
plt.show()
```

```python
sns.set(font_scale=1.5)                                                 # logistic- shows logistic regresiion.
sns.lmplot(data=titanic,x="age",y="survived",aspect=1,height=8,col=None,logistic=True) 
plt.show()
```

```python
sns.set(font_scale=1.5)                                                 
sns.lmplot(data=titanic,x="age",y="survived",aspect=1,height=8,col="sex",logistic=True) 
plt.show()
```

```python
sns.set(font_scale=1.5)                                                 
sns.lmplot(data=titanic,x="age",y="survived",aspect=1,height=8,col="pclass",logistic=True) 
plt.show()
```

## MATRIXPLOT/HEATMAPS

```python
titanic.head()
```

```python
pd.crosstab(titanic.sex,titanic.pclass) #Split the group and identify the size of each group.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass),annot=False,fmt="d",cmap=None,vmax=None)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass),annot=True,fmt="d",cmap="Reds",vmax=None)
plt.show()
```

```python
pd.crosstab(titanic.sex,titanic.pclass,values=titanic.survived,aggfunc="mean") # sggfunc=Means show mean table.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass,values=titanic.survived,aggfunc="mean"),annot=False,cmap="Reds")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass,values=titanic.survived,aggfunc="mean"),annot=True,cmap="Reds")
plt.show()
```

```python
titanic.corr()  # define a correlation matrix
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(titanic.corr(),annot=True,cmap="Reds")
plt.show()
```
